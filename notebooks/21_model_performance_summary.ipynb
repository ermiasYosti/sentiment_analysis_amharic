Eighteen models were run during the model selection process for Twitter sentiment analysis.  A baseline model was created that used a basic Naive Bayes Classifier with word vectors.  Basic Baseline Models are also included for reference that include picking one of the three classes or just random guessing.

The models were judged by the following metrics in descending order of importance:
* Macro Average Recall (*Macro Recall* above)
* Balanced Recall for each of the three classes
* Overall Accuracy

The model that was selected as the best was a Naive Bayes Classifier that was tuned using Min Doc Frequency, used tf/idf, SMOTE oversampling and lemmatization.  It performed marginally better than the same model but without lemmatization.  Its performance metrics were:
* Macro Average Recall 0.61
* Per Class Recall
    * Negative 0.61
    * Neutral 0.56
    * Positive 0.67
* Accuracy 0.60
